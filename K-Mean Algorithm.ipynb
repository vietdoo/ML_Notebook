```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.datasets.samples_generator import make_blobs 
import matplotlib.image as mpimg
import cv2
from IPython.display import clear_output
from time import sleep
import matplotlib.image as mpimg
import cv2
from IPython.display import clear_output
from time import sleep
def dis(p1, p2):
  return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
```

```python
cenx = [[], [], [], [], []]
ceny = [[], [], [], [], []]
k = 5
center = [[-4, -4], [7, 3], [6, 4]]#, [-4, 6], [-5, 3]]
colors = ['red', 'green', 'yellow', 'blue', 'pink']
x, y = make_blobs(n_samples=1000, centers=[[4,4], [-2, -1], [2, -3], [1, 1], [-2, 4]], cluster_std=0.85)
plt.rcParams['figure.figsize'] = [12, 6] 
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.scatter(x[:, 0], x[:, 1], marker='.')
for i in range(len(center)):
  ax.scatter(center[i][0], center[i][1], color = 'white', marker='o',  s = 1200, edgecolors='black' ,linewidth=6)  

plt.show()
for j in range(50):
  cenx = [[], [], [], [], []]
  ceny = [[], [], [], [], []]
  for point in x:
    d = 999
    c = 0
    for i in range(len(center)):
      if (d > dis(center[i], point)):
        d = dis(center[i], point)
        c = i
    cenx[c].append(point[0])
    ceny[c].append(point[1])
    
  for i in range(len(center)):
    center[i][0] = np.mean(cenx[i])
    center[i][1] = np.mean(ceny[i])
 
for i in range(len(center)):
  plt.scatter(cenx[i], ceny[i], color = colors[i], marker='.', s = 350, edgecolors='black')
  plt.scatter(center[i][0], center[i][1], color = 'white', marker='o',  s = 1500, edgecolors='black' ,linewidth=8)  

plt.show()
```
